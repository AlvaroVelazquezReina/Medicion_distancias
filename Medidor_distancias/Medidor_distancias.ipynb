```python
import math
```

```python
start_pos = [0,0]
end_pos = [0,0]
```

```python
actions = [('UP', 5),
          ('DOWN', 3),
          ('LEFT', 3),
          ('RIGHT', 2)]
```

```python
for action in actions:
    ## get the direction and steps
    direction = action[0]
    steps     = action[1]
    
    if direction == 'UP':
        end_pos[1] += steps
    elif direction == 'DOWN':
        end_pos[1] -= steps
    elif direction == 'LEFT':
        end_pos[0] -= steps
    elif direction == 'RIGHT':
        end_pos[0] += steps
    else:
        pass
```

## EUCLIDEAN DISTANCE

```python
## redefinimos variables (que estaban arriba)
start_pos = [1,3]
end_pos = [2, 5]
```

```python
def euclidean_distance(end_pos, start_pos):
    
    dist = math.sqrt((end_pos[0] - start_pos[0])**2 + (end_pos[1] - start_pos[1])**2)
    
    return dist
```

```python
print('Euclidean distance is = {}'.format( euclidean_distance(end_pos, start_pos)))
```

## Manhattan distance:

```python
## redefinimos variables (que estaban arriba)
start_pos = [1,3]
end_pos = [2, 5]
```

```python
def manhattan_distance (end_pos, start_pos):
    
    dist = abs(start_pos[0] - end_pos[0]) + abs(start_pos[1] - end_pos[1])
    
    return dist
```

```python
print('Manhattan distance is = {}'.format( manhattan_distance(start_pos, end_pos)))
```

## Minkowski distance:

```python
def minkowski_distance(end_pos, start_pos, root):
    
    dist = ((end_pos[0] - start_pos[0])**root + (end_pos[1] - start_pos[1])**root) ** (1/root)
             ## LO QUE VARIA 'X'  'mas porque es valor absoluto' lo que varia 'Y'
    
    return dist
```

```python
root = 1
print('Minkowski distance with root = {}, is = {}'.format(root, minkowski_distance(end_pos, start_pos, root)))

root = 2
print('Minkowski distance with root = {}, is = {}'.format(root, minkowski_distance(end_pos, start_pos, root)))

root = 3
print('Minkowski distance with root = {}, is = {}'.format(root, minkowski_distance(end_pos, start_pos, root)))
```

## Chebyshev distance:

```python
## recordamos las variables
start_pos = [5,5]
end_pos = [20, 8]
```

```python
def chebyshev_distance (end_pos, start_pos):
    
    dist = max(abs(start_pos[0] - end_pos[0]), abs(start_pos[1] - end_pos[1]))
    
    return dist
```

```python
print('Chebyshev distance is = {}'.format(chebyshev_distance(end_pos, start_pos)))
```
